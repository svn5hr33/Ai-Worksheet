```python
#Prove A.A−1 = I
import numpy as np
A = np.array([[3,4],[7,8]], dtype = int)
B = np.array([[5,3],[2,1]], dtype = int)
A_inv = np.linalg.inv(A)
result = np.dot(A, A_inv)
result_int = np.round(result).astype(int)
print(f"A * A^-1 = \n {result_int}")
print("\n")
```

```python
#Prove AB ̸= BA
import numpy as np
A = np.array([[3,4],[7,8]], dtype = int)
B = np.array([[5,3],[2,1]], dtype = int)
result_ab= np.dot(A, B)
result_ba= np.dot(B, A)

print(f"AB = \n {result_ab}")
print(f"BA = \n {result_ba}")
print("Since the value of AB and BA are different, therefore AB ̸= BA.")
print("\n")
```

```python
#Prove (AB)T = BTAT
import numpy as np
A = np.array([[3,4],[7,8]], dtype = int)
B = np.array([[5,3],[2,1]], dtype = int)
AB_T = np.transpose(np.dot(A, B))
BT_AT = np.dot(np.transpose(B), np.transpose(A))

print("(AB)^T =\n", AB_T)
print("B^T A^T =\n", BT_AT)
print("Since (AB)^T and B^TA^T are same therefore (AB)^T = B^TA^T ")
print("\n")
```

```python
#4.Solve the following system of Linear equation using Inverse Methods.
#2x − 3y + z = −1
#x − y + 2z = −3
#3x + y − z = 9
import numpy as np
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

B_eq = np.array([-1, -3, 9])
x = np.linalg.solve(A_eq, B_eq).astype(int)
print("This is using only np.linalg only ")
print(x)
print("\n")
#Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function of Numpy}
a_inv = np.linalg.inv(A_eq)
x = np.dot(a_inv, B_eq).astype(int)
print("This is using np.linagl.inv function ")
print(x)
```
